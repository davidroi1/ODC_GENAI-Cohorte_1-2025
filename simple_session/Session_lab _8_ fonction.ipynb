On veut crypter une chaine de caractères données `message` dont la taille ne dépasse pas 50 caractères en une chaine résultat `message_encoded` de la manière suivante :
- On parcours la chaine `message` de gauche à droite en comptant le nombre d'occurrence successives de chaque caractère de la chaine.
- Générez ensuite une nouvelle chaîne `message_encoded`, où chaque caractère est précédé du nombre de fois où il apparaît successivement.

Ecrire une fonction Python qui prend en entrée une chaine `message` qui doit être non vide de taille 50 maximum et formée uniquement par des lettres alphabétiques, puis retourne la chaine `message_encoded` selon le principe décrit précédemment.

Indications:
- Lever les exceptions nécessaires pour garantir le respect des conditions ci-dessus.
- Documenter la fonction en respectant les bonnes pratiques.

Exemple:
- Si message = "aaaFyBssssssssssssazz" alors la chaine message_encoded = "3a1F1y1B12s1a2z"

```python
# code ici
from collections import Counter as counter

message = "aaaFyBssssssssssssazz"


def string_encoder(value_:dict) -> str:
  encoded = ''
  for key, val in value_.items():
    encoded += ''.join((str(val), key))

  return encoded


def string_counter(string_:str) -> dict:
  try:
    exception_verification(string_)
  except Exception as e:
    return "erreur: la variable est vide ou taille superieur a 50 ou contient des caracteres special"
  else:
    string_dict = counter(string_)
    return string_dict


def exception_verification(string_:str) -> bool:
  if string_ == '' or len(string_) > 50 or not string_.isalnum():
    raise ValueError("erreur: la variable saisite est vide ou a une valeur superieur a 50 caracteres ou contient des caractere special")

  return True

response = string_counter(message)
encoding_response = string_encoder(response)
print(response)
print(encoding_response)

```
